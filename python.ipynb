```python
## POSITIONAL ARGUMENTS

def fun (*args):
   for nn in args:
    print(nn)
fun(1,2,3,4)
```

```python
## KEYWORDS ARGUMENTSS
def fun(**kwargs):   # use **kwargs for keyword arguments
    for key, value in kwargs.items():   # use .items()
        print(key, ":", value)

fun(name="Sabana", age=24, title="SK")
```
