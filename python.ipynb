```python
## Shallow Copy **********
# d_copy=d
# print("d_copy ", d_copy)
# print("d ", d)

# d_copy['age']=23
# print("d_copy ", d_copy)
# print("d ", d)


##### Shallow Copy

d_copy=d.copy()
print("d_copy ", d_copy)
print("d ", d)

d_copy['age']=23
print("d_copy ", d_copy)
print("d ", d)
```

```python
## Different types of iteration in dictionary

for keys in d.keys():
  print(keys)
  print (keys, ':', d[keys])

for values in d.values():
  print(values)

for key, value in d.items():
  print(f"{key}: {value}")
```

```python
##### Nested Dictionary
students = {
    "Alice": {"age": 23, "major": "Computer Science"},
    "Bob": {"age": 21, "major": "Mathematics"},
    "Charlie": {"age": 22, "major": "Physics"}
}

students
```

```python
students.keys()
students.values()
```

```python
for name, details in students.items():
    print(f"Student: {name}")
    for key, value in details.items():
        print(f"  {key}: {value}")
```
